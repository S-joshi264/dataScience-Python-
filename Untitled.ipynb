```python
def load_data(file):
    with open(file,"r") as f:
        data=json.load(f)
    return data
data=(load_data("C:\\all_codes\\dataScience(Python)\\facebook_users_dirty.json"))
def display_data(data):
    for i in data:
        print(f"User ID:{i["user_id"]} {i["name"]} is freind with {i["friends"]} who liked pages {i["liked_pages"]} ")
def cleanData(data):
    for c,i in enumerate(data):
        for j in list(i.keys()): 
            if i[j] is None:
                i.pop(j)
        if 'age' in i.keys() and int(i['age'])<0:
            i.pop('age')
def people_might_know(user,data):
    direct=[]
    for c, i in enumerate(data): 
        if  i["user_id"]==user :
            break
        
        if c==len(data)-1:
            return "user id not found"   
    for c, i in enumerate(data):   
        if i["user_id"]==user:
            for j in i['friends']:
                direct.append(j)
    suggested=[]
    for i in direct:
        for j in (data):
            if j["user_id"]==i:
                for  k in j["friends"]:
                    if k not in direct and k!=user  :
                    
                        suggested.append(k)
                
    return suggested
for j in (data):
    print(people_might_know(j["user_id"],data))
    
data=(load_data("C:\\all_codes\\dataScience(Python)\\facebook_users_dirty.json"))
cleanData(data)
print(display_data(data))
# display_data(cleanDatadata)
```
